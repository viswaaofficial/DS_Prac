```python
import pandas as pd

# Sample dataset (replace this with your actual dataset)
data = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Income': [50000, 60000, 75000, 80000, 90000],
    'Score': [85, 88, 90, 92, 95]
})

# 1. Count the number of non-null values in each column
count = data.count()
print("Count:")
print(count)

# 2. Calculate the mean (average) of numeric columns
mean = data.mean()
print("\nMean:")
print(mean)

# 3. Calculate the median of numeric columns
median = data.median()
print("\nMedian:")
print(median)

# 4. Calculate the standard deviation of numeric columns
std = data.std()
print("\nStandard Deviation:")
print(std)

# 5. Find the minimum and maximum values in numeric columns
min_val = data.min()
max_val = data.max()
print("\nMinimum values:")
print(min_val)
print("\nMaximum values:")
print(max_val)

# 6. Calculate the sum of values in numeric columns
sum_values = data.sum()
print("\nSum:")
print(sum_values)

# 7. Calculate the variance of numeric columns
variance = data.var()
print("\nVariance:")
print(variance)

# 8. Compute the correlation between columns
correlation = data.corr()
print("\nCorrelation:")
print(correlation)

# 9. Calculate specific percentiles of numeric columns
percentiles = data.quantile([0.25, 0.5, 0.75])
print("\nPercentiles:")
print(percentiles)

# 10. Get a summary of descriptive statistics for all numeric columns
summary = data.describe()
print("\nSummary:")
print(summary)

# 11. Count the unique values in a column
value_counts = data['Age'].value_counts()
print("\nValue Counts:")
print(value_counts)

# 12. Calculate the skewness and kurtosis of numeric columns
skewness = data.skew()
kurtosis = data.kurt()
print("\nSkewness:")
print(skewness)
print("\nKurtosis:")
print(kurtosis)
```

```python
import pandas as pd

# Sample DataFrame with multiple columns
data = pd.DataFrame({
    'Date': ['2022-01-01', '2022-01-01', '2022-01-02', '2022-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Value1': [10, 20, 15, 25],
    'Value2': [30, 40, 35, 45]
})

# Pivot the DataFrame based on 'Date' and 'Category' columns, with multiple value columns
pivot_data = data.pivot(index='Date', columns='Category', values=['Value1', 'Value2'])

# Display the pivoted DataFrame
print(pivot_data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
data = np.random.randn(100)

# Sort the data for the quantile plot
data.sort()

# Create a figure with two subplots (box plot and quantile plot)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Box Plot
ax1.boxplot(data)
ax1.set_xlabel('Data')
ax1.set_ylabel('Values')
ax1.set_title('Box Plot')

# Quantile Plot
quantiles = np.arange(0, 1.1, 0.1)
quantile_values = np.percentile(data, quantiles * 100)
ax2.plot(quantiles, quantile_values, marker='o', linestyle='--')
ax2.set_xlabel('Quantiles')
ax2.set_ylabel('Values')
ax2.set_title('Quantile Plot')
ax2.grid(True)

plt.tight_layout()
plt.show()
```
