```python
import numpy as np
import matplotlib.pyplot as plt

# Define the proportions of black and non-black individuals in the eligible population
eligible_population = [0.26, 0.74]

# Number of trials (simulations)
num_simulations = 10000

# Number of jurors in each panel
panel_size = 100

# Initialize an array to store the results of simulations
simulated_counts = []

# Simulate the process
for _ in range(num_simulations):
    # Randomly sample from the eligible population
    simulated_panel = np.random.choice([0, 1], size=panel_size, p=eligible_population)
    # Count the number of black men in the simulated panel
    black_count = np.sum(simulated_panel)
    simulated_counts.append(black_count)

# Create a histogram to visualize the results
plt.hist(simulated_counts, bins=range(0, panel_size + 1), density=True, alpha=0.5, color='blue', label='Simulated Data')
plt.axvline(x=8, color='red', linestyle='dashed', label='Swain\'s Panel')
plt.xlabel('Number of Black Men in the Panel')
plt.ylabel('Density')
plt.legend()
plt.show()
```
